```python
def random_mean_imputation(df, column, method = "mean", range = 0.1) :

    #defining the option for replacing nulls
    if method == "mean" : 
        core_replace = df[column].mean()
    elif method == "median" : 
        core_replace = df[column].median()
    else :
        raise ValueError("Method must be 'mean' or 'median'")
    #setting the bounds
    min_value = core_replace * (1 - range)
    max_value = core_replace * (1 + range)

      # Apply imputation
    df[column] = df[column].apply(
        lambda x: np.random.uniform(min_value, max_value) if pd.isnull(x) else x)
    return df
```
